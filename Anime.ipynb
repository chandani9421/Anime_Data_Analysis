```python
# function to extract the number of episodes from the string
def extract_episodes(txt):
    check = False
    data = ""
    for i in txt:
      if i == ")":
         check == False
         return data
      if check == True:
           data = data + i
      if i == '(':
           check = True
```

```python
# extracting the number of episodes and creating a new colummn called "Episodes"
df["Episodes"] = df["Title"].apply(extract_episodes)
```

```python
# removing the " eps" from the "Episodes" column
df["Episodes"] = df["Episodes"].str.replace(" eps","")
```

```python
# coverting the data type of the "Episodes" column to integer
df["Episodes"] = df["Episodes"].astype(int)
```

```python
# function to extract the time of release from the string
def extraction_time(txt):
    data = ""
    for i in range(len(txt)):
        if txt[i] == ")":
           for j in range(i+1,i+20):
               data += txt[j]

           return data  
```

```python
# extracting the time of release and creating a new column called "Total time"
df["Total time"] = df["Title"].apply(extraction_time)
```
